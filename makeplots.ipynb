```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,1,0.05)
y1 = 1*np.exp(-x) 
y2 = 1*np.exp(-2*x)
```

```python
plt.plot(x,y1, label = 'c1 y1')
plt.plot(x,y2, label = 'c2 y2')
plt.plot(x,y1  + y2, label = 'y(t)')
plt.legend()

```

```python
x = np.arange(0,10,0.05)
y1 = 1*np.exp(-x) 
y2 = 2*y1*x
```

```python
plt.plot(x,y1, label = 'c1 y1')
plt.plot(x,y2, label = 'c2 y2')
plt.plot(x,y1  + y2, label = 'y(t)')
plt.legend()
```
